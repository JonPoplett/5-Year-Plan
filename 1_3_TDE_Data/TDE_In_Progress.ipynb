```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Adjusted Constants
omega_complexity = 1 # Adjusted for complexity component
omega_countercomplexity = 0.5  # Ajusted for counter-complexity component
g = 0.13  # Coupling constant for interaction
N = 35
tlist = np.linspace(0, 540, 540)

# Operators
a = tensor(destroy(N), qeye(N))
adag = a.dag()
b = tensor(qeye(N), destroy(N))
bdag = b.dag()

# Position Operators
sigma_x = tensor(position(N), qeye(N))
sigma_y = tensor(qeye(N), position(N))

# Hamiltonian
H = omega_complexity * adag * a + omega_countercomplexity * bdag * b + g * (adag * b + a * bdag)

# Adjusted Initial State
alpha = 4.17
# Adjusted to target the desired magnitude range
psi0 = tensor(coherent(N, alpha), basis(N, 0))

# Expectation operators
e_ops = [adag * a, bdag * b, omega_complexity * adag * a, omega_countercomplexity * bdag * b, H, g * (adag * b + a * bdag), sigma_x, sigma_y]

# Time Evolution
output = mesolve(H, psi0, tlist, [], e_ops)

# Extracting energy components and position expectation values
E_complexity = output.expect[2]
E_countercomplexity = output.expect[3]
E_interaction = output.expect[5]
E_total = output.expect[4]
exp_sigma_x = output.expect[6]
exp_sigma_y = output.expect[7]

# Data storage
# We organize our simulation results into a DataFrame for easier analysis and visualization, akin to cataloging TDE observations.
data = {
    'Time': tlist,
    'Luminosity (Photon number)': output.expect[0],
    'E_complexity': E_complexity,
    'E_countercomplexity': E_countercomplexity,
    'E_interaction': E_interaction,
    'E_total': E_total,
    'Sigma_x': exp_sigma_x,
    'Sigma_y': exp_sigma_y
}
df = pd.DataFrame(data)

# Data export
# Save the simulation data to a CSV file, providing us with a record of the system's behavior over time.
filename = "quantum_simulation_data.csv"
df.to_csv(filename, index=False)
print(f"Data saved to {filename}")

# Visualization
# We plot various aspects of the system over time to visualize the evolution of complexity, counter-complexity, and their interaction.

# Plotting the photon number over time to represent the "luminosity" akin to observing the light curve of a TDE.
plt.figure(figsize=(10, 6))
plt.plot(tlist, output.expect[0], label='Photon number')
plt.xlabel('Time')
plt.ylabel('Photon number expectation value')
plt.title('Luminosity Over Time')
plt.legend()
plt.show()

# Plotting the energy distributions over time, giving us insights into the exchange of energy between complexity and counter-complexity components.
plt.figure(figsize=(12, 7))
plt.plot(tlist, E_complexity, label='E_complexity')
plt.plot(tlist, E_countercomplexity, label='E_countercomplexity')
plt.plot(tlist, E_interaction, label='E_interaction')
plt.plot(tlist, E_total, label='E_total', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Energy Distributions Over Time')
plt.legend()
plt.show()

# Plotting the position expectation values over time, allowing us to observe the "motion" of complexity and counter-complexity.
plt.figure(figsize=(12, 7))
plt.plot(tlist, exp_sigma_x, label='Sigma_x')
plt.plot(tlist, exp_sigma_y, label='Sigma_y')
plt.xlabel('Time')
plt.ylabel('Position expectation value')
plt.title('Photon Position Expectation Values Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import linear_sum_assignment
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

# Load model data
df_model = pd.read_csv("quantum_simulation_data.csv")
interaction = df_model['E_interaction'].values
counter = df_model['Luminosity (Photon number)'].values

# Load experimental data
df_exp = pd.read_csv("sorted_magnitude_data.csv")
change_in_intensity_exp = np.sort(df_exp['Magnitude'].values)

# Calculate combined cost for all three metrics
cost_matrix = np.zeros((len(counter), len(change_in_intensity_exp)))

for i in range(len(counter)):
    for j in range(len(change_in_intensity_exp)):
        cost_e = np.abs(counter[i] - change_in_intensity_exp[j])

        
        # Combining the costs
        total_cost = cost_e
        
        cost_matrix[i, j] = total_cost

# Use linear sum assignment to find the optimal assignment
row_idx, col_idx = linear_sum_assignment(cost_matrix)

# Sort the experimental data based on the optimal assignment
sorted_change_in_intensity = change_in_intensity_exp[col_idx]

# Compute statistics for E_countercomplexity and sorted_change_in_intensity
mae = mean_absolute_error(counter, sorted_change_in_intensity)
rmse = np.sqrt(mean_squared_error(counter, sorted_change_in_intensity))
r2 = r2_score(counter, sorted_change_in_intensity)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Square Error (RMSE): {rmse}")
print(f"R² (Correlation Coefficient): {r2}")

# Plotting can be similar to before, but now you have more variables. For simplicity, only the E_countercomplexity vs sorted_change_in_intensity is shown here:
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(counter, label="Model Data (E_complexity)", color="blue")
plt.title("Model Data for E_countercomplexity")
plt.xlabel("Data Points")
plt.ylabel("E_countercomplexity Value")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(sorted_change_in_intensity, label="Optimally Sorted Experimental Data (Change in Intensity)", color="red")
plt.title("Sorted Experimental Data for Change in Intensity")
plt.xlabel("Data Points")
plt.ylabel("Change in Intensity Value")
plt.legend()

plt.tight_layout()
plt.show()

# If you want to compare the two on the same plot:
plt.figure(figsize=(12, 6))
plt.plot(counter, label="Model Data (E_countercomplexity)", color="blue")
plt.plot(sorted_change_in_intensity, label="Optimally Sorted Experimental Data (Change in Intensity)", color="red")
plt.title("Comparison of Model vs Experimental Data")
plt.xlabel("Data Points")
plt.ylabel("Values")
plt.legend()
plt.show()

# Save the sorted experimental data and X_Center, Y_Center into the model dataframe
df_model["sorted_change_in_intensity"] = sorted_change_in_intensity
df_model.to_csv("quantum_simulation_data.csv", index=False)  # Saving the updated dataframe back to the csv

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
df = pd.read_csv('quantum_simulation_data.csv')

# Define the feature set and target variable
features = ['sorted_change_in_intensity']
target = 'E_complexity'

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print model evaluation metrics
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Print model coefficients
coefficients = pd.DataFrame(model.coef_, features, columns=['Coefficient'])
print(coefficients)

# Print y-intercept
print(f'Y-Intercept: {model.intercept_}')
```
