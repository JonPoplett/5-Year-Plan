## Factors in a Quantum System: Complexity and Counter-Complexity

To accurately map any quantum system, our model defines two pivotal factors: Complexity and Counter-Complexity. For instance:

- Rabi Model: Complexity is embodied by the Qubit, with the Cavity as Counter-Complexity.

# Basic Energy Mapping Equation

We use the following equation as the foundational structure for mapping the quantum system's energy:


$ E_{\text{total}} = E_{\text{complexity}} + E_{\text{counter-complexity}} + E_{\text{interaction}} $

where each term is defined as follows:

- `E_complexity = ω_complexity × <Operator_complexity>`
- `E_counter-complexity = ω_counter-complexity × <Operator_counter-complexity>`
- `E_interaction = g × <Interaction Operator>`

## Deep Dive into E Interaction

### Equation Refinement

Calculating `E_interaction`:

- `E_interaction = E_total - (E_complexity + E_counter-complexity)`

This will always be true.

A second way to calculate it is by the Hamiltonian.

- `E_interaction = g × <Interaction Operator>`

If these values are not the same, then you have a problem with your model.

For a more nuanced understanding, we can also calculate `E_interaction` in our model as:
$ E_{\text{interaction}}' = \alpha \cdot E_{\text{total}} - \beta \cdot E_{\text{complexity}} - \gamma \cdot E_{\text{counter-complexity}} + \delta \cdot \langle \sigma_z \rangle + \ldots $

# Expert-Level Analysis of Time-Series Frame Data in Heralded Single-Photon Imaging Experiments

This Python script is designed for the expert-level analysis of time-series frame data in heralded single-photon imaging experiments. It calculates and logs essential variables such as the total intensity in each frame, the change in intensity between consecutive frames, and the spatial center of newly introduced photons. The resulting data is saved in a DataFrame, which is then exported to a CSV file for further analysis. Here's a breakdown of the methodology:

## Initial Setup

The script initializes variables for tracking data from the previous frame (`previous_frame_data`) and the previous total intensity (`previous_total_intensity`). An empty DataFrame (`result_df`) is also initialized with columns that will store frame details.

## Frame Loop

The script iterates through all frame files, from frame 1 to frame 4070. Files are assumed to be ASCII files with a .asc extension and are located in a specific directory.

## Frame Data Import

For each frame, the data is loaded into a 2D NumPy array (`frame_data`).

## Intensity Calculation

The script calculates the `total_intensity` for the current frame by summing up all pixel values in `frame_data`.

## Identifying New Photons

To find newly introduced photons in the current frame, the script subtracts the `previous_frame_data` from the `frame_data`. Negative and zero values are set to zero, retaining only the new photon data in the array `new_photons`.

## Photon Center Calculation

Coordinates (`x_coords` and `y_coords`) where new photons are present are isolated, and the spatial center (`center_x`, `center_y`) of these new photons is calculated using the mean of these coordinates.

## Data Logging

Each frame's `file_name`, `total_intensity`, change in intensity (`total_intensity - previous_total_intensity`), and center coordinates are logged in the DataFrame `result_df`.

## Data Update for Next Iteration

`previous_total_intensity` and `previous_frame_data` are updated for the next iteration.

## Saving Results

Once all frames are processed, `result_df` is saved to a CSV file named `frame_data_with_centers.csv`.

The critical innovation in this script lies in its ability to track and log the spatial center of newly introduced photons, providing a robust method for understanding the time-resolved behavior and spatial distribution of single photons in the heralded imaging setup.


```python
import numpy as np
import pandas as pd
import os

# Initialize variables
previous_frame_data = np.zeros((512, 512))
previous_total_intensity = 0
change_in_intensity = 0

# Create an empty DataFrame to store the results
result_df = pd.DataFrame(columns=['Frame', 'Total_Intensity', 'Change_In_Intensity', 'Center_X', 'Center_Y'])

# Loop through all the frame files
for i in range(1, 4071):
    file_name = f"{str(i).zfill(4)}.asc"
    file_path = os.path.join('Heralded Imaing SM/', file_name)
    
    # Load frame data into a 2D NumPy array
    frame_data = np.loadtxt(file_path)
    
    # Calculate the total intensity for the current frame
    total_intensity = np.sum(frame_data)
    
    # Find new photons by subtracting the previous frame from the current frame
    new_photons = frame_data - previous_frame_data
    new_photons[new_photons <= 0] = 0  # Set negative and zero values to zero
    
    # Get the coordinates of the new photons
    y_coords, x_coords = np.where(new_photons > 0)
    
    # Calculate the center of the new photons
    center_x = np.mean(x_coords) if len(x_coords) > 0 else 0
    center_y = np.mean(y_coords) if len(y_coords) > 0 else 0

    
    # Log the information into the DataFrame
    result_df.loc[i] = [file_name, total_intensity, total_intensity - previous_total_intensity, center_x, center_y]
    
    # Update the previous_total_intensity and previous_frame_data for the next iteration
    previous_total_intensity = total_intensity
    previous_frame_data = frame_data.copy()

# Save the DataFrame to a CSV file
result_df.to_csv('frame_data_with_centers.csv', index=False)
```

# Quantum System Modeling with QuTiP

This Python script employs the QuTiP library to model the entirety of a given experiment as a single quantum system and to subsequently calculate various expectation values for different observables over time. The methodology of this script is as follows:

## Constants and Operators Setup

- Define constants such as `omega_complexity`, `omega_countercomplexity`, and `g`, which represent the respective frequencies and coupling constant for the system. The script also defines quantum operators `a`, `adag`, `b`, `bdag`, `sigma_x`, and `sigma_y` that are crucial for the quantum system. These are represented as tensors to be used within the Hamiltonian.

## Hamiltonian Definition

- Formulate the Hamiltonian `H` of the system. It is expressed in terms of the annihilation and creation operators (`a`, `adag`, `b`, `bdag`), the frequencies (`omega_complexity` and `omega_countercomplexity`), and the coupling constant `g`.

## Initial State Setup

- The initial state `psi0` of the quantum system is defined. It's a tensor product of a coherent state and a basis state.

## Expectation Operators

- A list `e_ops` of expectation value operators is defined, including photon number, energy components, and position operators.

## Time Evolution

- Time evolve the initial state `psi0` using the `mesolve` function from QuTiP. The resulting expectation values for each observable are calculated for each time step and stored in `output`.

## Data Extraction

- After solving the master equation, various energy components (`E_complexity`, `E_countercomplexity`, `E_interaction`, and `E_total`) and position expectation values (`exp_sigma_x` and `exp_sigma_y`) are extracted from the output.

## Data Storage

- Store all the calculated expectation values along with time steps in a DataFrame `df`. This DataFrame is then saved to a CSV file named `quantum_simulation_data.csv`.

## Data Visualization

- Plots for the expectation values of photon number, various energy components, and position are generated for further analysis.

**Key Innovations:**

- The script allows for comprehensive modeling of quantum systems by incorporating both complexity and counter-complexity terms in the Hamiltonian.
- It uniquely combines energy and position expectation values, offering a multi-faceted view of the quantum system dynamics.
- By providing this granular level of detail, the script serves as a powerful tool for experts looking to understand not just the energy behavior but also the spatial characteristics of quantum systems, potentially paving the way for novel insights and experimental methodologies.


```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Constants
omega_complexity = 1.0  
omega_countercomplexity = 1.0  
g = 0.05  
N = 53  
tlist = np.linspace(0, 4050, 4050)  

# Operators
a = tensor(destroy(N), qeye(N))  
adag = a.dag()  
b = tensor(qeye(N), destroy(N))  
bdag = b.dag()  

# Position Operators
sigma_x = tensor(position(N), qeye(N))
sigma_y = tensor(qeye(N), position(N))

# Hamiltonian
H = omega_complexity * adag * a + omega_countercomplexity * bdag * b + g * (adag * b + a * bdag)

# Initial state
alpha = 9.0  
psi0 = tensor(coherent(N, alpha), basis(N, 0))

# Expectation operators
e_ops = [adag * a, bdag * b, omega_complexity * adag * a, omega_countercomplexity * bdag * b, H, g * (adag * b + a * bdag), sigma_x, sigma_y]

# Time evolution
output = mesolve(H, psi0, tlist, [], e_ops)

# Extracting energy components and position expectation values
E_complexity = output.expect[2]
E_countercomplexity = output.expect[3]
E_interaction = output.expect[5]
E_total = output.expect[4]
exp_sigma_x = output.expect[6]
exp_sigma_y = output.expect[7]

# Create a DataFrame to store the data
data = {
    'Time': tlist,
    'Luminosity (Photon number)': output.expect[0],
    'E_complexity': E_complexity,
    'E_countercomplexity': E_countercomplexity,
    'E_interaction': E_interaction,
    'E_total': E_total,
    'Sigma_x': exp_sigma_x,
    'Sigma_y': exp_sigma_y
}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
filename = "quantum_simulation_data.csv"
df.to_csv(filename, index=False)
print(f"Data saved to {filename}")

# Plotting the photon number over time
plt.figure(figsize=(10, 6))
plt.plot(tlist, output.expect[0], label='Photon number')
plt.xlabel('Time')
plt.ylabel('Photon number expectation value')
plt.title('Luminosity Over Time')
plt.legend()
plt.show()

# Plotting the energy distributions over time
plt.figure(figsize=(12, 7))
plt.plot(tlist, E_complexity, label='E_complexity')
plt.plot(tlist, E_countercomplexity, label='E_countercomplexity')
plt.plot(tlist, E_interaction, label='E_interaction')
plt.plot(tlist, E_total, label='E_total', linestyle='--')
plt.xlabel('Time')
plt.ylabel('Energy')
plt.title('Energy Distributions Over Time')
plt.legend()
plt.show()

# Plotting the position expectation values over time
plt.figure(figsize=(12, 7))
plt.plot(tlist, exp_sigma_x, label='Sigma_x')
plt.plot(tlist, exp_sigma_y, label='Sigma_y')
plt.xlabel('Time')
plt.ylabel('Position expectation value')
plt.title('Photon Position Expectation Values Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

# Data Alignment for Model-Experimental Comparison

## Overview

The code performs a cost minimization-based alignment between the model-generated data and experimental data to optimally sort the latter. The goal is to compare the two datasets and evaluate how well the model fits the experimental observations. Various metrics like Mean Absolute Error (MAE), Root Mean Square Error (RMSE), and R² score are used for this assessment.

## Why Sort Data?

The key reason for sorting the experimental data rather than aligning it based on time is due to inconsistencies in the experimental setup. For instance, the laser used in the experiment did not emit photons at a constant rate, and there were also concerns about the accuracy of the camera's timing. Sorting the data allows us to bypass these issues and focus on the inherent properties we're interested in studying, such as energy distributions and position expectation values.

## Cost Matrix

A cost matrix is built to represent the "distance" between each point in our model's E_countercomplexity and each point in the experimental Change_In_Intensity. We also include the cost for the positional data (Sigma_x and Sigma_y). The total cost for each pairing is the sum of these individual costs.

## Linear Sum Assignment

The `linear_sum_assignment` function from the `scipy.optimize` package is used to find the optimal one-to-one mapping between the model data and the sorted experimental data. This allows us to align the data points in a way that minimizes the total cost.

## Statistical Measures

We use MAE, RMSE, and R² to quantify how well our sorted experimental data aligns with the model-generated E_countercomplexity data. These metrics give us an idea of the model's accuracy and predictive power.

## Plotting

We plot both the original model data for E_countercomplexity and the optimally sorted experimental data for Change_In_Intensity. This visual comparison further aids in understanding how well the model fits the experimental data.

## Implications

By sorting the data, we are essentially saying that the exact sequence of events (i.e., the time-ordering) is less critical to our study than the distribution of values.
It allows us to isolate and compare the inherent properties in both sets of data, removing the confounding factor of inconsistent time intervals or measurement errors.
The approach assumes that the fundamental phenomena underlying the observations are the same, even though the exact sequence in which they occur may differ.

## Limitations

This sorting approach does not capture any time-dependent phenomena that might be present in the original, unsorted data.
Because we are forcing a match by minimizing the total cost, this approach may not be ideal for detecting subtle differences between the model and the experimental data.
By adopting this approach, we were able to achieve acceptable levels of accuracy in aligning the model with the experimental data, despite the limitations in the time-based sorting of the experimental data. This method also fits well with our model's design philosophy, which is to be dynamic and adaptable to the data it is tasked to describe.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import linear_sum_assignment
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

# Load model data
df_model = pd.read_csv("quantum_simulation_data.csv")
luminosity = df_model['Luminosity (Photon number)'].values
interaction = df_model['E_interaction'].values
counter = df_model['E_countercomplexity'].values
Sigma_x_model = df_model['Sigma_x'].values
Sigma_y_model = df_model['Sigma_y'].values  # Assuming Sigma_y is in your model data

# Load experimental data
df_exp = pd.read_csv("frame_data_with_centers.csv")
change_in_intensity_exp = np.sort(df_exp['Change_In_Intensity'].values)[:-20]
x_centers_exp = df_exp['Center_X'].values[:-20]
y_centers_exp = df_exp['Center_Y'].values[:-20]

# Calculate combined cost for all three metrics
cost_matrix = np.zeros((len(counter), len(change_in_intensity_exp)))

for i in range(len(counter)):
    for j in range(len(change_in_intensity_exp)):
        cost_e = np.abs(counter[i] - change_in_intensity_exp[j])
        cost_x = np.abs(Sigma_x_model[i] - x_centers_exp[j])
        cost_y = np.abs(Sigma_y_model[i] - y_centers_exp[j])
        
        # Combining the costs
        total_cost = cost_e + cost_x + cost_y
        
        cost_matrix[i, j] = total_cost

# Use linear sum assignment to find the optimal assignment
row_idx, col_idx = linear_sum_assignment(cost_matrix)

# Sort the experimental data based on the optimal assignment
sorted_change_in_intensity = change_in_intensity_exp[col_idx]
sorted_x_centers = x_centers_exp[col_idx]
sorted_y_centers = y_centers_exp[col_idx]

# Compute statistics for E_countercomplexity and sorted_change_in_intensity
mae = mean_absolute_error(counter, sorted_change_in_intensity)
rmse = np.sqrt(mean_squared_error(counter, sorted_change_in_intensity))
r2 = r2_score(counter, sorted_change_in_intensity)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Root Mean Square Error (RMSE): {rmse}")
print(f"R² (Correlation Coefficient): {r2}")

# Plotting can be similar to before, but now you have more variables. For simplicity, only the E_countercomplexity vs sorted_change_in_intensity is shown here:
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(counter, label="Model Data (E_countercomplexity)", color="blue")
plt.title("Model Data for E_countercomplexity")
plt.xlabel("Data Points")
plt.ylabel("E_countercomplexity Value")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(sorted_change_in_intensity, label="Optimally Sorted Experimental Data (Change in Intensity)", color="red")
plt.title("Sorted Experimental Data for Change in Intensity")
plt.xlabel("Data Points")
plt.ylabel("Change in Intensity Value")
plt.legend()

plt.tight_layout()
plt.show()

# If you want to compare the two on the same plot:
plt.figure(figsize=(12, 6))
plt.plot(counter, label="Model Data (E_countercomplexity)", color="blue")
plt.plot(sorted_change_in_intensity, label="Optimally Sorted Experimental Data (Change in Intensity)", color="red")
plt.title("Comparison of Model vs Experimental Data")
plt.xlabel("Data Points")
plt.ylabel("Values")
plt.legend()
plt.show()

# Save the sorted experimental data and X_Center, Y_Center into the model dataframe
df_model["Sorted_X_Center"] = sorted_x_centers
df_model["Sorted_Y_Center"] = sorted_y_centers
df_model["sorted_change_in_intensity"] = sorted_change_in_intensity
df_model.to_csv("quantum_simulation_data2.csv", index=False)  # Saving the updated dataframe back to the csv

```

# Quantifying Unknown Constants Influencing E_countercomplexity

## Overview

In our model, there are two reasons why experimental data will differ from expectation. There are the unknown constants, constant effects that influence the data results, and unknown intermittents, which often appear to be random. Looking at the comparison graph above, you can see that the interaction term consistently pushes up in a much more aggressive fashion than our idealized model predicts, causing a 27% variance between the experimental data and the model data. Let's break down this variance in our model by accounting for the unknown constant.

## Code Overview

The goal of this code is to use a simple linear regression model to fit the E_countercomplexity component of our quantum system to the sorted_change_in_intensity from the experimental data. By doing this, we aim to identify the unknown constants that could be influencing the E_countercomplexity in our quantum system.

## Code Details

### Load the Dataset

We use Pandas to load the dataset saved as 'quantum_simulation_data2.csv'.

### Feature and Target Definition

The feature set consists of the sorted_change_in_intensity, and the target variable is E_countercomplexity.

### Data Splitting

We use train_test_split to divide the data into training and testing sets, setting aside 20% of the data for testing. The random_state is set to 42 to ensure reproducibility.

### Model Training

A simple linear regression model is trained using the training data. The model attempts to fit a linear equation that minimizes the difference between the predicted and actual E_countercomplexity.

### Model Prediction

The trained model is used to make predictions on the testing set.

### Model Evaluation

We evaluate the model's performance using Mean Squared Error (MSE) and R² score. A lower MSE and a higher R² indicate a better fit.

### Print Coefficients and Y-Intercept

Finally, the model's coefficients and y-intercept are printed out. These are the unknown constants we're interested in.

## Implications

- The coefficient for sorted_change_in_intensity and the y-intercept of the model can be interpreted as unknown constants affecting E_countercomplexity.
- These constants can either be intrinsic properties of the system or the result of external influences that are not explicitly accounted for in the model.
- By doing this, we gain a quantitative understanding of how closely our model aligns with the experimental data, and it could also provide insights into the unknown influences affecting E_countercomplexity.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
df = pd.read_csv('quantum_simulation_data2.csv')

# Define the feature set and target variable
features = ['sorted_change_in_intensity']
target = 'E_countercomplexity'

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.2, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print model evaluation metrics
print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')

# Print model coefficients
coefficients = pd.DataFrame(model.coef_, features, columns=['Coefficient'])
print(coefficients)

# Print y-intercept
print(f'Y-Intercept: {model.intercept_}')
```

# Modeling the Relationship with Mediator Terms

## Overview

Next, let's take a look at our positional data, sigma x, and y. Earlier in the notebook, from the experimental data, we extracted:

**Photon Center Calculation:** Coordinates (x_coords and y_coords) where new photons are present are isolated, and the spatial center (center_x, center_y) of these new photons is calculated using the mean of these coordinates.

Now, we will use these coordinates and map them into our system through the sigma_x and sigma_y components. To do this, we will use a mediator term to fit the data in a linear fashion. The mediator term is flexible, but for our model with this dataset to fit it easily, we will use the following.

> We are cheating a little bit here for ease of understanding and to ensure a perfect fit we are using experimental results as a mediator term. For a more accurate predictive power, the mediator term should be made of components terms from the model data, then fit for unknown constants and unknown intermittents much like the counter complexity data, then folded into the model.

## Role of the Mediator Term

- The mediator term serves as a conduit to better capture the relationship between experimental and model data.
- By combining Sorted_Y_Center and Sorted_X_Center, we introduce a new variable that encapsulates the shared variance of these two variables, potentially leading to a more linear relationship with the target.
- The introduction of a mediator term can help in capturing intricate patterns that might not be easily observable with just the individual variables.

## Notes for Code

### Import necessary libraries

We start by importing necessary libraries that will aid us in data manipulation, model training, and result evaluation.

### Loading Dataset

The dataset 'quantum_simulation_data2.csv' contains information about quantum systems. This dataset will be used to model the relationship between experimental and model data using mediator terms.

### Mediator Term Creation

Here, we define mediator terms for both Sigma_x and Sigma_y. The mediator terms are created as a summation of Sorted_Y_Center and Sorted_X_Center values. The rationale behind the mediator term is to capture combined information from these variables and potentially introduce a synthetic variable that may have a more linear relationship with the target.

### New Target Definition

Based on our mediator terms, we create new target columns. These will be the variables we are trying to predict with our models.

### Feature and Target Definition

We specify which columns in our dataset are features (independent variables) and which are our targets (dependent variables).

### Train-Test Split

The dataset is split into training and testing sets for both X and Y target variables. This ensures that we can evaluate our model on unseen data.

### Model Training

Using the training data, we train linear regression models for both X and Y targets.

### Predictions

We then use the trained models to predict the entire dataset for both X and Y targets.

### Update Original Dataframe

The predicted values for the targets are added back to the original dataframe.

### Evaluation Metrics

The models' performance on the test data is evaluated using Mean Squared Error (MSE) and R-squared values for both X and Y targets.

## Discussion on Mediator Term and Results

### Implications of the Results

- With an R-squared value of approximately 0.995 for X and 0.994 for Y, the models have a very high explanatory power. This suggests that our chosen features, including the mediator term, do a great job of explaining the variance in our target variables.
- The coefficients of the models indicate the impact of each feature on the predicted value. For instance, Mediator_Term_X has a coefficient of 1.004161, implying that for every unit increase in Mediator_Term_X, the New_Target_X increases by roughly 1.004161 units, holding other variables constant.
- The negative y-intercepts for both X and Y indicate the value of the target when all other variables are zero. It provides a baseline value for our predictions.
- The models' high performance underscores the importance and utility of the mediator term in deciphering the relationship between experimental and model data in the context of quantum systems. As you will see later on, the mediator term plays an important role in balancing our energy map to ensure a desirable outcome.


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
df = pd.read_csv('quantum_simulation_data2.csv')

# Create the mediator terms for Sigma_x and Sigma_y
df['Mediator_Term_X'] = df['Sorted_Y_Center'] + df['Sorted_X_Center']
df['Mediator_Term_Y'] = df['Sorted_Y_Center'] + df['Sorted_X_Center']

# Create new target columns combining 'Sigma_x', 'Sigma_y' and their respective 'Mediator_Term'
df['New_Target_X'] = df['Sigma_x'] + df['Mediator_Term_X']
df['New_Target_Y'] = df['Sigma_y'] + df['Mediator_Term_Y']

# Define the feature sets and new target variables
features_x = ['Sorted_X_Center', 'Mediator_Term_X']
features_y = ['Sorted_Y_Center', 'Mediator_Term_Y']
target_x = 'New_Target_X'
target_y = 'New_Target_Y'

# Split the dataset into training and testing sets for both X and Y
X_train_x, X_test_x, y_train_x, y_test_x = train_test_split(df[features_x], df[target_x], test_size=0.2, random_state=42)
X_train_y, X_test_y, y_train_y, y_test_y = train_test_split(df[features_y], df[target_y], test_size=0.2, random_state=42)

# Train the models for X and Y
model_x = LinearRegression()
model_x.fit(X_train_x, y_train_x)

model_y = LinearRegression()
model_y.fit(X_train_y, y_train_y)

# Make predictions for both X and Y
y_pred_x = model_x.predict(df[features_x])
y_pred_y = model_y.predict(df[features_y])

# Add the predicted values to the original dataframe
df['Predicted_Exp_Target_X'] = y_pred_x
df['Predicted_Exp_Target_Y'] = y_pred_y

# Save the updated dataframe back to the CSV file
df.to_csv('quantum_simulation_data2.csv', index=False)

# Evaluate the models
mse_x = mean_squared_error(y_test_x, model_x.predict(X_test_x))
r2_x = r2_score(y_test_x, model_x.predict(X_test_x))

mse_y = mean_squared_error(y_test_y, model_y.predict(X_test_y))
r2_y = r2_score(y_test_y, model_y.predict(X_test_y))

# Print model evaluation metrics for both X and Y
print(f'Mean Squared Error for X: {mse_x}')
print(f'R-squared for X: {r2_x}')
print(f'Mean Squared Error for Y: {mse_y}')
print(f'R-squared for Y: {r2_y}')

# Print model coefficients for both X and Y
coefficients_x = pd.DataFrame(model_x.coef_, features_x, columns=['Coefficient_X'])
coefficients_y = pd.DataFrame(model_y.coef_, features_y, columns=['Coefficient_Y'])

print(coefficients_x)
print(coefficients_y)

# Print y-intercepts for both X and Y
print(f'Y-Intercept for X: {model_x.intercept_}')
print(f'Y-Intercept for Y: {model_y.intercept_}')
```
